### QUESTION 7 -
Normalement, vous avez complété les programmes précédents avec les distances que vous avez mesurées. Vérifiez que vous les avez tous executés, ou vous aurez une erreur ! Lancez l'algorithme suivant, qui trace la courbe du logarithme de la distance totale en fonction de la distance moyenne entre deux points. Que constatez-vous ?

```python
##################################### QUESTION 7 #############################################


#Affichage

plt.rcParams['figure.figsize'] = [14, 14] #Redimensionne les figures

a, b = np.polyfit(np.log(distance_moyenne_entre_deux_points), np.log(distance_totale), 1) #Calcul de la droite la plus proche de la mesure


plt.plot(distance_moyenne_entre_deux_points, distance_totale, label='Distance totale mesurée')

# Affiche la grille
plt.grid(True, which="both")

# Titre
plt.title('Distance totale mesuree entre St Brieuc et le Mont Saint-Michel en fonction de la distance moyenne entre deux points')

# Axe x
plt.xlabel('distance moyenne entre deux points')
 
# Axe y
plt.ylabel('Distance totale mesuree')

# Affichage
plt.show()




fig, ax = plt.subplots()

ax.plot(np.log(distance_moyenne_entre_deux_points), np.log(distance_totale), label='Mesure') #Affiche les points mesurés
ax.plot(np.log(distance_moyenne_entre_deux_points), np.multiply(np.log(distance_moyenne_entre_deux_points),a)+b, label='Droite interpolee (a='+str(round(a,3))+')') #Affiche la droite interpolée



# Affiche la grille
plt.grid(True, which="both")

# Titre
plt.title('log(distance totale) en fonction de log(distance moyenne entre deux points)')

# Axe x
plt.xlabel('log(distance moyenne entre deux points)')

# Axe y
plt.ylabel('log(Distance totale mesuree)')

legend = ax.legend(loc='upper left', shadow=True, fontsize='x-large')


# Affichage
plt.show()
```

On constate que la courbe de log(distance_totale) en fonction de log(distance_moyenne_entre_deux_points) est proche d'une droite ! On peut montrer que son coefficient directeur a est tel que D=1+|a| ou D est la dimension fractale de cette portion de côte. Avec nos valeurs, on trouve donc une dimension fractale de la côte de 1.19. Cette dimension est comprise entre 1 et 2 : la forme de la côte n'est ni une courbe de dimension 1, ni une surface de dimension 2, mais "quelque chose" entre les deux : une fractale ! De plus cette dimension nous apprend des choses intéressante sur la forme de la côte : plus D est proche de 1, plus la côte est régulière et sa géométrie se rapproche d'une simple courbe. À l'inverse, plus D est proche de 2, plus la côte est irrégulière et se rapproche d'une surface.
